```python
class LogNormWrapper:
    def fit(self,data):
        print(inspect.getfullargspec(stats.lognorm.fit))
        return stats.lognorm.fit(data=data,s=1)
    def pdf(self,*args):
        print(*args[0])
        return stats.lognorm.pdf(*args)
wrapper = LogNormWrapper()    
sns.distplot(preprocessing.scale(train_data['SalePrice'].astype(float)), hist=False , kde=True,rug=True,fit=wrapper)
```
