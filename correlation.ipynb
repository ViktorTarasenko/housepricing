```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
```

```python
all_data = pd.read_csv('train_cluster_str.csv')
```

```python
sns.set(font_scale=0.7)
all_data = pd.get_dummies(all_data)
corr = all_data[all_data.SalePrice>1].corr()
top_corr_cols = corr[abs((corr.SalePrice)>=.26)].SalePrice.sort_values(ascending=False).keys()
top_corr = corr.loc[top_corr_cols, top_corr_cols]
dropSelf = np.zeros_like(top_corr)
dropSelf[np.triu_indices_from(dropSelf)] = True
plt.figure(figsize=(20, 20))
sns.heatmap(top_corr, cmap=sns.diverging_palette(880, 10, as_cmap=True), annot=True, fmt=".2f", mask=dropSelf)
plt.show()
```
